```python
lista_vendas=[1000,500,800,1500,2000,2300]
meta = 1300
contador=0

for venda in lista_vendas:
  if venda > meta:
    #contador = contador +1
    contador +=1
print('quantidade de vendas que atingiram a meta',contador)
```

```python
lista_vendas = [1000, 500, 800, 1500, 2000, 2300]
meta = 1200
percentual_extra = 0.1
bonus_extra = 50

for venda in lista_vendas:
  bonus=0

  if venda>=meta:
    bonus = venda * percentual_extra

    if venda>=2000:
      bonus+=bonus_extra

  print(f"venda:{venda}->bonus:R${bonus}")
```
